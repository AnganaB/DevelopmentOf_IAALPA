```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("svm_train.csv")
```

```python
df.columns
```

```python
df=df.drop(columns=['Unnamed: 0','aai_pair'])
```

```python
X=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
X
```

```python
from sklearn.svm import SVC
```

```python
clf=SVC(kernel='rbf')
```

```python
model=clf.fit(X.values,y.values)
```

```python
model
```

```python
import pickle
```

```python
pickle.dump(model,open("svm.pkl","wb"))
```
