```python
#importing different libraries

import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
from getAAI import s1,s2,s3,s4,s5,s6,s7,s8,s9,s10,s11,s12,s13,s14,s15,s16,s17,s18,s19,s20,s21,s22,s23,s24,s25
from auc import getNetworkAUC, getBestAUC, getComplementaryAAI, functionList, getEdgelist
```

```python
#load the edgelist from the csv file 

n = pd.read_csv("Dataset/network_edgelist.csv")
n.head()
```

```python
#create a networkx graph from the edgelist and display

network = nx.from_pandas_edgelist(n, source="Source", target="Target", create_using=nx.DiGraph()).to_undirected()
nx.draw_spring(network, with_labels=True)
```

```python
#generate attention allocation indicies for all connected node pairs in the graph

fn_list = functionList()
edge_list = getEdgelist(network)

nw_aai = []

for e in edge_list:
    e_aai = []
    
    for fn in fn_list:
        e_aai.append(eval(fn)(e,network))
    nw_aai.append(e_aai)

data = {
    'Source':[e[0] for e in edge_list],
    'Target':[e[1] for e in edge_list],
    'S1':[e_aai[0] for e_aai in nw_aai],
    'S2':[e_aai[1] for e_aai in nw_aai],
    'S3':[e_aai[2] for e_aai in nw_aai],
    'S4':[e_aai[3] for e_aai in nw_aai],
    'S5':[e_aai[4] for e_aai in nw_aai],
    'S6':[e_aai[5] for e_aai in nw_aai],
    'S7':[e_aai[6] for e_aai in nw_aai],
    'S8':[e_aai[7] for e_aai in nw_aai],
    'S9':[e_aai[8] for e_aai in nw_aai],
    'S10':[e_aai[9] for e_aai in nw_aai],
    'S11':[e_aai[10] for e_aai in nw_aai],
    'S12':[e_aai[11] for e_aai in nw_aai],
    'S13':[e_aai[12] for e_aai in nw_aai],
    'S14':[e_aai[13] for e_aai in nw_aai],
    'S15':[e_aai[14] for e_aai in nw_aai],
    'S16':[e_aai[15] for e_aai in nw_aai],
    'S17':[e_aai[16] for e_aai in nw_aai],
    'S18':[e_aai[17] for e_aai in nw_aai],
    'S19':[e_aai[18] for e_aai in nw_aai],
    'S20':[e_aai[19] for e_aai in nw_aai],
    'S21':[e_aai[20] for e_aai in nw_aai],
    'S22':[e_aai[21] for e_aai in nw_aai],
    'S23':[e_aai[22] for e_aai in nw_aai],
    'S24':[e_aai[23] for e_aai in nw_aai],
    'S25':[e_aai[24] for e_aai in nw_aai],
}

network_aai = pd.DataFrame(data)
network_aai
```
