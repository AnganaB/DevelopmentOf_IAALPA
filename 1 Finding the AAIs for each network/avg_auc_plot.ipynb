```python
import pandas as pd
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
from io import StringIO 
from auc import getNetworkAUC
import math
```

```python
#selecting a few networks to compute the AUC of different AAIs
network_auc = []
networks = ["126067398.edges","258955239.edges","58525541.edges","45145944.edges","207594668.edges",
           "27595997.edges", "48730516.edges", "227785045.edges"]

for network in networks:
    n = pd.read_csv("Dataset/"+network, sep=" ", names=['Source', 'Target'])
    csv_string = StringIO(n.to_csv(None, sep=","))
    auc_list = getNetworkAUC(csv_string)
    network_auc.append(auc_list)
network_auc
```

```python
#Average AUC is computed for each AAI
size = len(networks)
avg_auc = []

for i in range(25):
    s = sum(row[i] for row in network_auc)
    avg_auc.append(s/size)

min_val = math.floor(min(avg_auc)*10)/10 - 0.1
max_val = math.ceil(max(avg_auc)*10)/10 + 0.1
```

```python
#plot of avg AUC per AAI

x = ['S1', 'S2', 'S3', 'S4', 'S5', 'S6', 'S7', 'S8', 'S9', 'S10', 'S11', 'S12', 'S13', 'S14', 'S15', 'S16', 
    'S17', 'S18', 'S19', 'S20', 'S21', 'S22', 'S23', 'S24', 'S25']

plt.figure(num=None, figsize=(16, 6))
plt.plot(x,avg_auc, marker='o')
plt.yticks(np.arange(min_val, max_val, step=0.1))
plt.xlabel("AAI")
plt.ylabel("AUC")
```
